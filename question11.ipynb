```python
import matplotlib.pyplot as plt

def plot(data):
    """
    Plot 2D binary-class data.
    Each element in `data` should be [x1, x2, label].
    """
    data = list(data)
    x1 = [d[0] for d in data]
    x2 = [d[1] for d in data]
    y  = [d[2] for d in data]

    # separate classes
    class0_x1 = [x1[i] for i in range(len(y)) if y[i] == 0]
    class0_x2 = [x2[i] for i in range(len(y)) if y[i] == 0]
    class1_x1 = [x1[i] for i in range(len(y)) if y[i] == 1]
    class1_x2 = [x2[i] for i in range(len(y)) if y[i] == 1]

    plt.figure(figsize=(5,5))
    plt.scatter(class0_x1, class0_x2, color='red', label='Class 0', s=80, edgecolors='k')
    plt.scatter(class1_x1, class1_x2, color='blue', label='Class 1', s=80, edgecolors='k')
    plt.xlabel("Input A")
    plt.ylabel("Input B")
    plt.title("Logic-Gate Data Distribution")
    plt.legend()
    plt.grid(True)
    plt.show()
data = data_not_a_and_b
plot(data)
```

```python
import random

N  = 100

y_k = 45
custom_datas = []
for i in range(N) :
    if i%2 == 0 :
        x1 = random.randint(5, 100)
        x2 = random.randint(5,y_k)
        y = 0
    else :
        x1 = random.randint(5,100)
        x2 = random.randint(y_k, 100-y_k)
        y = 1

    custom_datas.append([x1,x2,y])

print(custom_datas)
plot(custom_datas)
```

Perceptron

```python
import matplotlib.pyplot as plt


w1 = 0.0
w2 = 0.0
b1 = 0.0

lr = 0.001
epochs = 500


data = custom_datas
print(data)
plot(data)

accuracies= []


for epoch in range(epochs) :

    step = 1
    corrects = 0
    total = len(data)

    for example in data :

        x1 = example[0]
        x2 = example[1]
        y_true = example[2]

        z = x1*w1 + x2*w2 + b1
        a = step_function(z)
        y_pred = a

        error = y_true - y_pred

        if error == 0 :
            corrects = corrects + 1

        w1 = w1 + lr * error * x1
        w2 = w2 + lr * error * x2
        b1 = b1 + lr * error

    accuracy = corrects / total
    accuracies.append(accuracy)

    # print("Epoch : {}, Accuracy : {}".format(epoch+1, accuracy))

print(accuracies)
plt.plot(accuracies)
print("w1 : {}, w2 : {}, b1 : {}".format(w1, w2, b1))
```

Neuron : Sigmoid + MSE


```python
import matplotlib.pyplot as plt
import numpy as np



w1 = 0.0
w2 = 0.0
b1 = 0.0

lr = 0.0005
epochs = 500


data = custom_datas
print(data)
plot(data)

accuracies= []


def sigmoid(x) :
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)



for epoch in range(epochs) :

    step = 1
    corrects = 0
    total = len(data)

    for example in data :

        x1 = example[0]
        x2 = example[1]
        y_true = example[2]

        z = x1*w1 + x2*w2 + b1
        a = sigmoid(z)
        y_pred = 1  if a>=0.5 else 0

        error = a - y_true
        transfer_derivative = sigmoid_derivative(z)

        error_term = error * transfer_derivative

        if y_true == y_pred :
            corrects = corrects + 1

        w1 = w1 - lr * error_term * x1
        w2 = w2 - lr * error_term * x2
        b1 = b1 - lr * error_term

    accuracy = corrects / total
    accuracies.append(accuracy)

    # print("Epoch : {}, Accuracy : {}".format(epoch+1, accuracy))

print(accuracies[-10:])
plt.plot(accuracies)
print("w1 : {}, w2 : {}, b1 : {}".format(w1, w2, b1))
```

```python
import matplotlib.pyplot as plt
import numpy as np



w1 = 0.0
w2 = 0.0
b1 = 0.0

lr = 0.0005
epochs = 50000


data = custom_datas
print(data)
plot(data)

accuracies= []


def sigmoid(x) :
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)



for epoch in range(epochs) :

    step = 1
    corrects = 0
    total = len(data)

    for example in data :

        x1 = example[0]
        x2 = example[1]
        y_true = example[2]

        z = x1*w1 + x2*w2 + b1
        a = sigmoid(z)
        y_pred = 1  if a>=0.5 else 0

        error = a - y_true


        error_term = error

        if y_true == y_pred :
            corrects = corrects + 1

        w1 = w1 - lr * error_term * x1
        w2 = w2 - lr * error_term * x2
        b1 = b1 - lr * error_term

    accuracy = corrects / total
    accuracies.append(accuracy)

    # print("Epoch : {}, Accuracy : {}".format(epoch+1, accuracy))

print(accuracies[-10:])
plt.plot(accuracies)
print("w1 : {}, w2 : {}, b1 : {}".format(w1, w2, b1))
```

```python
import numpy as np
import matplotlib.pyplot as plt



# Generate x1 values
x1 = np.linspace(-1, 1, 100)

# Compute x2 from the decision boundary equation: w1*x1 + w2*x2 + b1 = 0
x2 = -(w1 * x1 + b1) / w2

# Plot decision boundary
plt.plot(x1, x2, 'r-', label='Decision Boundary')

# Add region shading (optional)
plt.fill_between(x1, x2, color='blue', alpha=0.1, label='y_pred = 1 region')
plt.fill_between(x1, x2, y2=max(x2)+10, color='orange', alpha=0.1, label='y_pred = 0 region')

# Plot setup
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.title('Perceptron Decision Boundary')
plt.grid(True)
plt.show()
```
