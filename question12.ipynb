```python
import numpy as np
import matplotlib.pyplot as plt
def error_term_1 (y_true, y_pred):
    return (y_pred - y_true)

x = np.linspace(0,1,100)
y_true = 1
y = [error_term_1(y_true, y_pred) for y_pred in x]
plt.figure()
plt.plot(x, y)
plt.title("Sigmoid + BCE")
plt.xlabel("y_pred")
plt.ylabel("Error Term")
plt.grid(True)
plt.show()
```

```python
def error_term (y_true, y_pred):
    return (y_pred - y_true)*y_pred*(1-y_pred)

x = np.linspace(0,1,100)
y_true = 1
y = [error_term(y_true, y_pred) for y_pred in x]  # âœ… fixed to call correct function
plt.figure()
plt.plot(x, y)
plt.title("Sigmoid + MSE (0 to 1)")
plt.xlabel("y_pred")
plt.ylabel("Error Term")
plt.grid(True)
plt.show()
```

```python
def error_term (y_true, y_pred):
    return (y_pred - y_true)

x = np.linspace(-1000,1000,100)
y_true = 1
y = [error_term(y_true, y_pred) for y_pred in x]
plt.figure()
plt.plot(x, y)
plt.title("Linear + MSE (-1000 to 1000)")
plt.xlabel("y_pred")
plt.ylabel("Error Term")
plt.grid(True)
plt.show()
```

```python
def error_term (y_true, y_pred):
    a1 = y_pred - y_true
    a2 = 1 if y_pred > 0 else 0
    return a1*a2

x = np.linspace(-1000,1000,100)
y_true = 1
y = [error_term(y_true, y_pred) for y_pred in x]
plt.figure()
plt.plot(x, y)
plt.title("ReLU + MSE (-1000 to 1000)")
plt.xlabel("y_pred")
plt.ylabel("Error Term")
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-5, 5, 200)
y_linear = x
y_relu = np.maximum(0, x)

plt.figure(figsize=(8,5))
plt.plot(x, y_linear, label='Linear Output Layer', color='blue')
plt.plot(x, y_relu, label='ReLU Output Layer', color='red')
plt.title('Output Layer Comparison: Linear vs ReLU')
plt.xlabel('Input (z)')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def d_sigmoid(x):
    s = sigmoid(x)
    return s * (1 - s)

x = np.linspace(-10, 10, 400)
y = sigmoid(x)
dy = d_sigmoid(x)

plt.figure(figsize=(8,5))
plt.plot(x, y, label='Sigmoid f(x)', color='blue')
plt.plot(x, dy, label="Sigmoid Derivative f'(x)", color='red', linestyle='--')
plt.title('Sigmoid Activation Function and its Gradient')
plt.xlabel('x')
plt.ylabel('Output / Gradient')
plt.legend()
plt.grid(True)
plt.show()
```
